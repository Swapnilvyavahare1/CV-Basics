```python
# crop
def viewImage(image, name_of_window):
    cv2.namedWindow(name_of_window, cv2.WINDOW_NORMAL)
    cv2.imshow(name_of_window, image)
    cv2.waitKey()
    cv2.destroyAllWindows()
```

```python
viewImage(image1, 'Sample Image')
```

```python
image1.shape
```

```python
cropped_image= image1[200:800, 200:800]
viewImage(cropped_image, 'Cropped_image')
```

```python
# Rotation
(h,w,d)= image1.shape
center= (w/2, h/2)
M= cv2.getRotationMatrix2D(center, 45, 1.0)
rotated= cv2.warpAffine(image1, M, (w,h))
viewImage(rotated, 'Rotated Image')
```
